```python
Q1 = tdata.quantile(0.25)
Q3 = tdata.quantile(0.75)
IQR = Q3 - Q1
print(Q1)
```

```python
print(Q3)
```

```python
print(IQR)
```

```python
tdata1= tdata[~((tdata < (Q1 - 1.5 * IQR)) |(tdata > (Q3 + 1.5 * IQR))).any(axis=1)]
print(tdata1.shape)
print(tdata.shape)
```

```python
tdata1.corr()
```

```python
plt.subplots(figsize=(16,8))
sns.heatmap(tdata1.corr(),annot=True,cmap="YlGnBu")
plt.show()
```

```python
model2=smf.ols("tdata1.pm~tdata1.ambient + tdata1.coolant + tdata1.motor_speed+ tdata1.stator_yoke + tdata1.profile_id+  tdata1.stator_tooth + tdata1.stator_winding+ tdata1.u_q+tdata1.u_d+tdata1.i_q+tdata1.i_d",data=tdata1).fit()
print(model2.params)
model2.summary()
```
